```python
import pandas as pd
import numpy as np
import requests
from dotenv import load_dotenv
import os
import requests
import time
load_dotenv()
```

```python
GH_PERSONAL_ACCESS_TOKEN = os.getenv("GH_PERSONAL_ACCESS_TOKEN")
```

```python
spam_issues_df = pd.read_pickle("../data/spam-issues-1717917295.9858763.pkl")
```

Fetch spam issues

```python
headers = {f"Authorization": "Bearer {GH_PERSONAL_ACCESS_TOKEN}"}
owner = "McKenna8"
repo = "Everything-you-can-imagine-is-real"
state = "open"
page = 1
per_page = 25
response = requests.get(f"https://api.github.com/repos/{owner}/{repo}/issues?state={state}&page={page}&per_page={page}")

json = response.json()
print(json)
```

```python
normalised = pd.json_normalize(json)
normalised_filtered = normalised.filter(items=["id", "url", "user.login", "title", "body", "labels"])
normalised_filtered.head()
```
